```python
import numpy as np
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load user_signup_data file as a pandas dataFrame

df = pd.read_csv('n26_v1.csv',index_col=0)
```

```python
# Check the first rows of user_signup_data file 

df.head(1)
```

```python
#drop userid column

df.drop(['userid'] , axis=1 , inplace = True)
```

```python
# new dataFrame in cases when TV Ad was True or 1


df_tv_ad = df[df['campaign']==1]
```

```python
# Correlation Matrix when TV Ad was True or 1

df_tv_ad.corr()
```

```python
# new dataFrame in cases when TV Ad was False or 0

df_no_tv_ad = df[df['campaign']==0]
```

```python
# Heatmap to visualize the correlation including all data

sns.heatmap(df.corr() ,cmap='YlGnBu',linecolor='white',linewidths=1 ,annot=True)
```

```python
# Heatmap to visualize the correlation of only data with NO TV AD 


sns.heatmap(df_no_tv_ad.corr() ,cmap='YlGnBu',linecolor='white',linewidths=1 ,annot=True)
```

```python
# Heatmap to visualize the correlation of only data with TV AD 

sns.heatmap(df_tv_ad.corr() ,cmap='YlGnBu',linecolor='white',linewidths=1 ,annot=True)
```
